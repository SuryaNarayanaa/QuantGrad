```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def f(x):
    return 3*x**2 - 2*x - 1
f(-3)
```

```python
xVals = np.arange(-5.0 , 5.0 , 0.1)
yVals = f(xVals)
plt.plot(xVals , yVals)
```

```python
h = 0.0001
def differentiate(f , x):
    return (f(x + h) - f(x))/h
derivative = [x for x in xVals if abs(differentiate(f, x)) < 0.2]
derivative
```
